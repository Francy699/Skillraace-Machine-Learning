```python
from collections import namedtuple
import numpy as np

# Define a namedtuple to represent a data point
Point = namedtuple('Point', ['area', 'perimeter', 'label'])

def calculate_entropy(data):
  """Calculates the entropy of a dataset."""
  labels = [point.label for point in data]
  label_counts = {label: labels.count(label) for label in set(labels)}
  entropy = 0
  for count in label_counts.values():
    if count > 0:
      p = count / len(labels)
      entropy -= p * np.log2(p)
  return entropy

def information_gain(parent_entropy, data_left, data_right):
  """Calculates the information gain from a split."""
  weight_left = len(data_left) / (len(data_left) + len(data_right))
  weight_right = len(data_right) / (len(data_left) + len(data_right))
  return parent_entropy - (len(data_left) / len(parent_entropy) * calculate_entropy(data_left) + weight_right * calculate_entropy(data_right))

def split_data(data, feature_index, threshold):
  """Splits the data based on a feature and threshold."""
  data_left = [point for point in data if point[feature_index] <= threshold]
  data_right = [point for point in data if point[feature_index] > threshold]
  return data_left, data_right

def create_decision_tree(data, features, max_depth=3):
  """Recursively builds a decision tree."""
  if len(data) == 0 or len(set([point.label for point in data])) == 1 or max_depth == 0:
    return plurality_value(data)  # Return the most common label

  parent_entropy = calculate_entropy(data)
  best_gain = 0
  best_feature_index = None
  best_threshold = None

  for feature_index in range(len(features)):
    feature = features[feature_index]
    thresholds = set([point[feature_index] for point in data])
    for threshold in thresholds:
      data_left, data_right = split_data(data, feature_index, threshold)
      gain = information_gain(parent_entropy, data_left, data_right)
      if gain > best_gain:
        best_gain = gain
        best_feature_index = feature_index
        best_threshold = threshold

  # Split the data based on the best feature and threshold
  left_subtree = create_decision_tree(split_data(data, best_feature_index, best_threshold)[0], features[:best_feature_index] + features[best_feature_index + 1:], max_depth-1)
  right_subtree = create_decision_tree(split_data(data, best_feature_index, best_threshold)[1], features[:best_feature_index] + features[best_feature_index + 1:], max_depth-1)
  return (features[best_feature_index], best_threshold, left_subtree, right_subtree)

def plurality_value(data):
  """Returns the most common label in the data."""
  label_counts = {label: data.count(label) for label in set([point.label for point in data])}
  return max(label_counts, key=label_counts.get)

def classify(tree, point):
  """Classifies a new data point using the decision tree."""
  feature, threshold, left_subtree, right_subtree = tree
  if point[features.index(feature)] <= threshold:
    return classify(left_subtree, point)
  else:
    return classify(right_subtree, point)

# Sample data (replace with your actual data)
data = [
    Point(12, 10, "circle"),
    Point(15, 12, "circle"),
    Point(6, 12, "square"),
    Point(8, 16, "square"),
    Point(20, 14, "circle"),
    Point(10, 14, "square"),
]

features = ["area", "perimeter"]  # Feature names (replace with your features)

# Create the decision tree
tree = create_decision_tree(data,features,max_depth=3)

print('Decision Tree')
visualize_tree(tree)
```
